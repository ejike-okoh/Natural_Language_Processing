```python
import requests
from bs4 import BeautifulSoup
import gensim
import docx
```

```python
url = 'https://www.pau.edu.ng/news/professor-enase-resumes-as-the-deputy-vice-chancellor-of-pan-atlantic-university/'
```

```python
page = requests.get(url).text
page
```

```python
soup = BeautifulSoup(page)
```

```python
paragraph = soup.find_all('p')
```

```python
text = [tags.get_text().strip() for tags in paragraph]
```

```python
text
```

```python
sentence_list = [sentence for sentence in text if not '\n' in sentence]
```

```python
sentence_list = [sentence for sentence in sentence_list if '.' in sentence]
```

```python
sentence_list
```

```python
article = ''.join(sentence_list)
```

```python
doc = docx.Document()
```

```python
doc.add_picture('Prof.Enase-Okonedo.jpg')
```

```python
doc.add_paragraph(article)
```

```python
doc.save('PAU_New_DVC.docx')
```

```python
from gensim.summarization import summarize
```

```python
summarize(article)
```
